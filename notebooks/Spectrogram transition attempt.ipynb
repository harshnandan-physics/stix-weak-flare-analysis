```python
from stixdcpy import spectrogram as cspec
import numpy as np
import pandas as pd

pad = pd.Timedelta(minutes=10)
flare_start = pd.Timestamp("2021-09-25T09:12:00")
flare_end   = pd.Timestamp("2021-09-25T09:19:04")

spec = cspec.Spectrogram.from_sdc(
    (flare_start - pad).isoformat(),
    (flare_end + pad).isoformat(),
)

timestamp = pd.to_datetime(spec.data["datetime"])
timedel   = np.array(spec.data["timedel"])

# Boolean mask selecting bins that fall inside the catalog's actual flare interval
in_flare = (timestamp >= flare_start) & (timestamp <= flare_end)

print(f"Bins inside flare window: {in_flare.sum()} / {len(timestamp)}")
print("\ntimedel values INSIDE flare window:")
print(timedel[in_flare])
print("\ntimedel values in the 5 min immediately BEFORE flare start (background/pre window):")
pre_mask = (timestamp >= flare_start - pd.Timedelta(minutes=5)) & (timestamp < flare_start)
print(timedel[pre_mask])
```

```python
raw = spec.data["spectrogram"] * timedel[np.newaxis, :]   # broadcast timedel across channels

# If our hypothesis is right, every entry should be very close to an integer
frac_part = np.abs(raw - np.round(raw))
print("Max deviation from nearest integer across ALL channels/bins:", frac_part.max())
print("Mean deviation:", frac_part.mean())

# Spot check a handful of random bins
rng = np.random.default_rng(0)
sample_idx = rng.choice(raw.shape[1], size=5, replace=False)
print("\nSample raw-count columns (should look like clean integers):")
print(pd.DataFrame(raw[:, sample_idx], columns=[str(timestamp[i]) for i in sample_idx]).round(3))
```

```python
spec_day = cspec.Spectrogram.from_sdc("2021-09-25T00:02:00", "2021-09-25T23:30:00")

timedel_day = np.array(spec_day.data["timedel"])
counts_day  = spec_day.data["spectrogram"]          # shape (17, n_bins), rate units

# Total flux across all channels, per bin — a rough proxy for "how bright was the Sun in this bin"
total_flux_day = counts_day.sum(axis=0)

# If binning is flux-adaptive, timedel should shrink as total_flux_day grows.
# A quick correlation check (Spearman handles the non-linear relationship better than Pearson):
from scipy.stats import spearmanr
corr, pval = spearmanr(total_flux_day, timedel_day)
print(f"Spearman correlation (flux vs bin width): {corr:.3f}  (p = {pval:.2e})")

import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(total_flux_day, timedel_day, s=6, alpha=0.4)
plt.xlabel("Total flux across channels (counts/sec)")
plt.ylabel("timedel (s)")
plt.title("Does bin width shrink when flux rises?")
plt.show()
```

```python
import pandas as pd

flare_df = pd.read_csv("../data/processed/stix_flare_catalog_sep20_25_2021.csv")
flare_df["flare_id"] = flare_df["flare_id"].astype(str)

peak_cols = ["LC0_PEAK_COUNTS_4S","LC1_PEAK_COUNTS_4S","LC2_PEAK_COUNTS_4S","LC3_PEAK_COUNTS_4S","LC4_PEAK_COUNTS_4S"]

# Approx total flux proxy in counts/sec, matching the x-axis units from the correlation plot
flare_df["peak_flux_proxy"] = flare_df[peak_cols].sum(axis=1) / 4.0

print(flare_df["peak_flux_proxy"].describe())

# Rough threshold read off your scatter plot — bin width stays flat above this
threshold = 800
below = (flare_df["peak_flux_proxy"] < threshold).sum()
print(f"\nFlares with peak flux proxy BELOW ~{threshold} counts/sec: {below} / {len(flare_df)}  ({100*below/len(flare_df):.1f}%)")

# Where does our GOES B2.9 test flare rank?
test = flare_df[flare_df["flare_id"] == "2109250916"]
print("\nTest flare peak_flux_proxy:", test["peak_flux_proxy"].values)
print("Percentile rank in catalog:", (flare_df["peak_flux_proxy"] < test["peak_flux_proxy"].values[0]).mean() * 100)
```
