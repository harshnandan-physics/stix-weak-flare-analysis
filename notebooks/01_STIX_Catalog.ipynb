# 01 — STIX Flare Catalog

## Objective

Retrieve the STIX flare catalog for the observation period of Awasthi et al. (2021),
inspect its structure, and export a standardized processed catalog for use throughout
the pipeline.

## Inputs

- STIX Data Center (via `stixdcpy`)
- Observation interval: 2021-09-20 to 2021-09-25

## Outputs

- `flare_df` — pandas DataFrame, one row per detected flare (215 events)
- `stix_flare_catalog_sep20_25_2021.csv` — processed catalog saved to `data/processed/`

## Scientific Milestone

Establish the complete set of STIX flare detections for the study period.
All subsequent notebooks load from the exported CSV rather than querying the
STIX Data Center again.

## 1. Retrieve Flare Catalog

```python
from flare_pipeline.data import get_flare_catalog
```

```python
# Observation window from the paper
START_TIME = "2021-09-20T00:00:00"
END_TIME   = "2021-09-25T23:59:59"
```

```python
flares = get_flare_catalog(
    start_time=START_TIME,
    end_time=END_TIME,
)

print(f"Number of flares detected: {len(flares)}")
```

## 2. Convert and Export

Convert the raw API response to a pandas DataFrame and immediately export it
as the processed catalog. All subsequent notebooks load from this CSV — the
STIX Data Center is not queried again after this point.

```python
import pandas as pd

# Converting the list of flares into a dataframe. This is going to be our primary dataframe

flare_df = pd.DataFrame(flares)
```

```python
# Converting the table into a physical file to remove the dependance on online archive

flare_df.to_csv(
    "../data/processed/stix_flare_catalog_sep20_25_2021.csv",
    index = False
)
```

## 4. Data Quality — GOES Flux

Nine flares have `GOES_flux = 0`, corresponding to `GOES_class = NaN`.
These are not errors — GOES simply had no reliable flux estimate for those
events. They are retained in the catalog unchanged.

```python
flare_df["GOES_flux"].describe()
```

```python
# Count of flares with missing GOES flux
(flare_df["GOES_flux"] == 0).sum()
```

```python
# List the affected flares
flare_df.loc[
    flare_df["GOES_flux"] == 0,
    ["flare_id", "start_UTC", "peak_UTC", "GOES_class", "GOES_flux"]
]
```
